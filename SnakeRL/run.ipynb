```python
def env_creator(env_config):
    return SnakeEnv(env_config)
```

```python
snakie = SnakeEnv({'render_mode':"human"})
snakie.reset()
```

```python
#run forever and take actions from keyboard and collect data about the reward
while True:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            pygame.quit()
            quit()
        if event.type == pygame.KEYDOWN:
            if event.key == pygame.K_LEFT:
                snakie.snake.direction = (-1, 0)
            elif event.key == pygame.K_RIGHT:
                snakie.snake.direction = (1, 0)
            elif event.key == pygame.K_UP:
                snakie.snake.direction = (0, -1)
            elif event.key == pygame.K_DOWN:
                snakie.snake.direction = (0, 1)
    snakie.step(0)
    print(snakie.compute_reward(0))
    snakie.render("human")
    pygame.time.wait(100)
```
